```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/kc_house_data.csv')
```

```python
df.head()
```

```python
cols = ['id', 'date', 'price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'sqft_above', 'sqft_basement']

df = df[cols]
        
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df['waterfront'].value_counts()
```

```python
df['waterfront'] = df['waterfront'].fillna('NO')
```

```python
df.info()
```

```python
df['view'].value_counts()
```

```python
df['view'].isna().sum()
```

```python
df['view'] = df['view'].fillna('NONE')
```

```python
df.info()
```

```python
df['sqft_basement'].value_counts()
```

```python
df['sqft_basement'] = df['sqft_basement'].replace({'?':'0.0'})
```

```python
df['sqft_basement'] = df['sqft_basement'].astype('float')
```

```python
df.info()
```

```python
df['waterfront'].value_counts()
```

```python
df['waterfront'] = df['waterfront'].map({'NO':0, 'YES':1})
```

```python
df.info()
```

```python
df['view'].value_counts()
```

```python
df['view'] = df['view'].map({'NONE':0, 'FAIR':1, 'AVERAGE':2, 'GOOD':3, 'EXCELLENT':4})
```

```python
df.info()
```

```python
df.describe()
```

```python
corr = df.corr()
corr
```

```python
fig, ax = plt.subplots(figsize=(12,10)) 

sns.heatmap(corr, annot=True, ax=ax);
```
